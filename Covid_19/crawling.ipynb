```python
import pandas as pd 
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(executable_path="DATA/chromedriver.exe")
driver.get('https://www.worldometers.info/coronavirus/')
```

```python
id = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[1]")
id_ = id[:231]
country = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[2]")
country_ = country[:231]
total = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[3]")
total_ = total[:231]
newCase = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[4]")
newCase_ = newCase[:231]
totalDeath = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[5]")
totalDeath_ = totalDeath[:231]
newDeath = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[6]")
newDeath_ = newDeath[:231]
totalRecover = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[7]")
totalRecover_ = totalRecover[:231]
newRecover = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[8]")
newRecover_ = newRecover[:231]
activeCase = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[8]")
activeCase_ = activeCase[:231]
serious = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[10]")
serious_ = serious[:231]
TotCase = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[11]")
TotCase_ = TotCase[:231]
Deaths = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[12]")
Deaths_ = Deaths[:231]
totalTest = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[13]")
totalTest_ = totalTest[:231]
test = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[14]")
test_ = test[:231]
population = driver.find_elements(By.XPATH,"//table/tbody/tr[@class='even' or @class='odd']/td[15]")
population_ = population[:231]
```

```python
Col_name = ["ID","Country","Total Case", "New Case", "Total Deaths","New Deaths","Total Recovered","New Recovered","Active Case", "Serious","Tot Case","Deaths 1M", "Totals Tests","Tests 1M","Population"]
```

```python
ID = [e.text for e in id_] 
Country = [e.text for e in country_]
TotalCase = [e.text for e in total_]
NewCase = [e.text for e in newCase_]
TotalDeath = [e.text for e in totalDeath_]
NewDeaths = [e.text for e in newDeath_]
TotalRecovered = [e.text for e in totalRecover_]
NewRecovered = [e.text for e in newRecover_]
ActiveCase = [e.text for e in activeCase_]
Serious = [e.text for e in serious_]
TotCases = [e.text for e in TotCase_] 
Deaths_1M = [e.text for e in  Deaths_]
TotalTests = [e.text for e in totalTest_]
Tests = [e.text for e in test_]
Population =[e.text for e in population_]
```

```python
data = pd.DataFrame(list(zip(ID,Country, TotalCase, NewCase, TotalDeath, NewDeaths, TotalRecovered, NewRecovered, ActiveCase, Serious,TotCases,Deaths_1M,TotalTests,Tests, Population)),columns= Col_name)
```
