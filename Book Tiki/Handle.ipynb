1. Quyển sách nào bán chạy nhất.

```python
book= df.loc[df.all_time_quantity_sold == df['all_time_quantity_sold'].max(), ["name", "price"]]
book
```

2. Top 10 quyển sách bán chạy nhất và tổng danh thu

```python
feature =df[['name','all_time_quantity_sold', 'price']]
rsl2= feature.groupby('name').sum().sort_values('all_time_quantity_sold', ascending= False)
rsl2[:10]
```

```python
total_revenue = sum(rsl2[:10]['all_time_quantity_sold'] * rsl2[:10]['price'])
total_revenue
```

3. Top 10 quyển sách được giảm giá nhiều nhất. Cho biết giá niêm yết của nó.

```python
feature = df[['name', 'discount', 'price']]
feature['Listed price'] = feature['discount'] + feature['price']
rsl3= feature.sort_values('discount',ascending=False)[:10]
rsl3
```
