```python
import requests
from tqdm import tqdm
import pandas as pd
import time
import random
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36 Edg/110.0.1587.69',
    'Accept': 'application/json, text/plain, */*',
    'Accept-Language': 'vi-VN,vi;q=0.8,en-US;en;q=0.9',
    'Referer': 'https://tiki.vn/nha-sach-tiki/c8322',
    'x-guest-token': 'JB9w6xdmUh4kHlfYKi1yvzs8t3X5bTPr',
    'Connection': 'keep-alive',
    'TE': 'Trailers',
}

params = {
    'limit': '40',
    'include': 'advertisement',
    'aggregations': '2',
    'trackity_id': 'efa44528-9ee2-eb55-48b6-7d2dc608543d',
    'category': '8322',
    'page': '1',
    'urlKey':  'nha-sach-tiki',
}
```

```python
book_id = []
for i in range(1,21):
    params['page']= i
    clone = requests.get('https://tiki.vn/api/personalish/v1/blocks/listings', params= params, headers= headers)
    if clone.status_code == 200:
        for j in clone.json().get('data'):
            book_id.append({'ID': j.get('id')})
    time.sleep(3)
```

```python
p_id = pd.DataFrame(book_id)
p_id
```

```python

headers_detail = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36 Edg/110.0.1587.69',
    'Accept': 'application/json, text/plain, */*',
    'Accept-Language': 'vi-VN,vi;q=0.8,en-US;en;q=0.9',
    'Referer': 'https://tiki.vn/khong-diet-khong-sinh-dung-so-hai-tb5-p187827003.html?itm_campaign=CTP_YPD_TKA_PLA_UNK_ALL_UNK_UNK_UNK_UNK_X.213851_Y.1796171_Z.3592344_CN.AUTO---Khong-Diet-Khong-Sinh-%C4%90ung-So-Hai-%28TB5%29---2023%2F01%2F05-04%3A20%3A14&itm_medium=CPC&itm_source=tiki-ads&spid=187827005',
    'x-guest-token': 'JB9w6xdmUh4kHlfYKi1yvzs8t3X5bTPr',
    'Connection': 'keep-alive',
    'TE': 'Trailers',
}

params_detail = (
    ('platform', 'web'),
    ('spid', 187827005)
)


def parser_product(json):
    d = dict()
    d['id'] = json.get('id')
    d['sku'] = json.get('sku')
    d['name']= json.get('name')
    d['short_description'] = json.get('short_description')
    d['price'] = json.get('price')
    d['discount'] = json.get('discount')
    d['all_time_quantity_sold'] = json.get('all_time_quantity_sold')
    d['discount_rate'] = json.get('discount_rate')
    d['review_count'] = json.get('review_count')
    d['inventory_status'] = json.get('inventory_status')
    return d
```

```python
book_id_ = p_id.ID.to_list()
results_details= []
for i in tqdm(book_id_, total=len(book_id_)):
    clone_ = requests.get('https://tiki.vn/api/v2/products/{}'.format(i), params= params_detail, headers= headers_detail)
    if clone_.status_code == 200:
        results_details.append(parser_product(clone_.json()))
    time.sleep(4)
```

```python
results_ = pd.DataFrame(results_details)
results_
```

```python
results_.to_csv('BookDataTiki.csv', index= False)
```
